```python
# Confusion Matrices.
# Author : Udacity + kei
# Date   : February 4, 2018

# from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

X = pd.read_csv('data/titanic_data.csv')

X = X._get_numeric_data()
y = X['Survived']
del X['Age'], X['Survived']

from sklearn.tree import DecisionTreeClassifier
from sklearn import cross_validation # for version 0.17
X_train, X_test, y_train, y_test = cross_validation.train_test_split(
    X, y, test_size=0.25, random_state=0)

clf1 = DecisionTreeClassifier()
clf1.fit(X_train, y_train)
cm_DT = confusion_matrix(y_test, clf1.predict(X_test))
print "Confusion matrix for this Decision Tree:\n", cm_DT
```

```python
# view with a heatmap
sns.heatmap(cm_DT, annot=True, cmap='Blues', xticklabels=['no', 'yes'], yticklabels=['no', 'yes'])
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.title('Confusion matrix for:\n{}'.format(clf1.__class__.__name__));
```
